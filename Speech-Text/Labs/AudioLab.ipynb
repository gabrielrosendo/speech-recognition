```python
# AudioLab.ipynb

# Import libraries
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

# File path
file_path = 'cv-corpus-20.0-delta-2024-12-06/en/clips/common_voice_en_41236242.mp3'

# Load the audio file
x, sr = librosa.load(file_path)
print(f"Audio file: {file_path}")
print(f"Sample rate: {sr} Hz")
print(f"Duration: {len(x) / sr:.2f} seconds")

# Plot the waveform
plt.figure(figsize=(10, 4))
librosa.display.waveshow(x, sr=sr)
plt.title("Waveform of the Audio")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()

# Compute and plot the spectrogram
D = librosa.stft(x)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
plt.figure(figsize=(10, 6))
librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='log', cmap='viridis')
plt.colorbar(format='%+2.0f dB')
plt.title("Spectrogram (Log Scale)")
plt.show()

# Compute MFCCs
n_mfcc = 13

# Summary Findings
duration = len(x) / sr
peak_amplitude = np.max(np.abs(x))

print("\n### Summary of Findings ###")
print(f"Audio Duration: {duration:.2f} seconds")
print(f"Peak Amplitude: {peak_amplitude:.2f}")
```

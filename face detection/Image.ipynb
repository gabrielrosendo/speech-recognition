````python
import cv2
from matplotlib import pyplot as plt
````

````python
# Load the image
img = cv2.imread('face0.jpg')
````

````python
# Display the grayscale image using Matplotlib
plt.imshow(img)
plt.axis('off')  # Hide axis
plt.show()
````

````python
# Convert the image from BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
````

````python
# Display the grayscale image using Matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Hide axis
plt.show()
````

````python
# Load the Haar Cascade classifier for frontal face detection
cascade_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
````

````python
# Convert the image to grayscale for face detection
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
````

````python
# Convert the image from BGR to RGB
gray_rgb = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)
````

````python
# Display the grayscale image using Matplotlib
plt.imshow(gray_rgb)  # Set the colormap to gray
plt.axis('off')  # Hide axis
plt.show()
````

````python
# Detect faces in the image
faces = cascade_classifier.detectMultiScale(gray,minNeighbors=10)
````

````python
# Draw rectangles around detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 255, 0), 10)
    cv2.putText(img, 'face', (x + 10, y + 10), cv2.FONT_HERSHEY_SIMPLEX, 10, (255, 0, 255), 10)

````

````python
# Convert the image from BGR to RGB
final_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
````

````python
# Display the image with detected faces
# Display the grayscale image using Matplotlib
plt.imshow(final_rgb)  # Set the colormap to gray
plt.axis('off')  # Hide axis
plt.show()
````

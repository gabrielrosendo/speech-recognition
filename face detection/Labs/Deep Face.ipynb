```python
from deepface import DeepFace
from matplotlib import pyplot as plt
```

```python
# Detect Faces =======================================================
detected_face = DeepFace.detectFace("old_woman.webp")
print("Detected face:", detected_face)
```

```python
# Show Detected Face
plt.imshow(detected_face[0])
plt.axis('off')  # Hide axes
plt.title("Detected Face")  # Add title
plt.show()
```

```python
# Verify images ======================================================
result = DeepFace.verify("Selena.jpeg", "test.jpg")
print("Is verified: ", result["verified"])
```

```python
# Get age, gender, race, emotion =====================================
# DeepFace.analyze can perform multiple analyses on the input face
demography = DeepFace.analyze("old_lady.jpg")  # By default, it finds age, gender, race, and emotion
```

```python
# Access the first element of the list
demography = demography[0]
```

```python
# Print demographic information
print("Age: ", int(demography["age"]))
print("Emotion: ", demography["dominant_emotion"])
print("Gender: ", demography["gender"])
print("Race: ", demography["dominant_race"])
```

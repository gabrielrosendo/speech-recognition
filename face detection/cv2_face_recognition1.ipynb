```python
import cv2
from PIL import Image
import face_recognition
```

```python
image = cv2.imread("twilight.webp")
display(image)
```

```python
rgb_frame = image[:, :, ::-1]  #Convert to RGB format
display(rgb_frame)
```

```python
face_locations = face_recognition.face_locations(rgb_frame)
```

```python
count = 0
for face_location in face_locations:
    count = count + 1
    top, right, bottom, left = face_location
    print("Face {} Top: {}, Left: {}, Bottom: {}, Right: {}".format(count, top, left, bottom, right))

    face_image = rgb_frame[top:bottom, left:right]
    title = 'face' + str(count)
    #use cv2.imshow instead if you are running locally
    pil_image = Image.fromarray(face_image)
    display(title, pil_image)
```
